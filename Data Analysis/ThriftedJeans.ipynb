```python
# create another column that is the exponential moving average of the price
jeans['EMA15'] = jeans['Price'].ewm(span=15, adjust=False).mean()
jeans['EMA5'] = jeans['Price'].ewm(span=5, adjust=False).mean()
plot_column = jeans['EMA5']
# plot EMA5
plt.plot(jeans['Day'], jeans['Price'])
plt.plot(jeans['Day'], plot_column)
plt.xlim(0, 100)
```
